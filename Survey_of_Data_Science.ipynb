The following function given a word, returns a list of strings with one character deleted.

````
def delete_letter(word, verbose=False):
    '''
    Input:
        word: the string/word for which you will generate all possible words 
                in the vocabulary which have 1 missing character
    Output:
        delete_l: a list of all possible strings obtained by deleting 1 character from word
    '''
    
    delete_l = []
    split_l = []
    
    ### START CODE HERE ###
    split_l = [(word[:i],word[i:]) for i in range(len(word)+1)]
    delete_l = [word[:i]+word[i+1:] for i in range(len(word))]
    ### END CODE HERE ###

    if verbose: print(f"input word {word}, \nsplit_l = {split_l}, \ndelete_l = {delete_l}")

    return delete_l
````

Switch letter. A function that switches two letters in a word:

````
def switch_letter(word, verbose=False):
    ''' 
    Input:
        word: input string
     Output:
        switches: a list of all possible strings with one adjacent charater switched
    ''' 
    
    switch_l = []
    split_l = []

    split_l = [(word[:i],word[i:]) for i in range(len(word))]
    switch_l = [i + j[1] + j[0] + j[2:] for i,j in split_l if len(j) >= 2]      
    
    if verbose: print(f"Input word = {word} \nsplit_l = {split_l} \nswitch_l = {switch_l}") 

    return switch_l
````

##Counts

The implementation of a dictionary of word counts for a given text is a powerful tool.

````
from collections import Counter

def get_count(word_l):
    '''
    Input:
        word_l: a set of words representing the corpus. 
    Output:
        worid_count_dict: The wordcount dictionary where key is the word and value is its frequency.
    '''
    
    word_count_dict = {} 
    word_count_dic = Counter()
    for word in word_l: word_count_dic[word] += 1
    word_count_dict = dict(word_count_dic) 
    return word_count_dict
````

````
wordcount = get_count(articleWords)
print(wordcount)

from google.colab.output import eval_js
eval_js('google.colab.output.setIframeHeight("500")')
````

Now, given the dictionary of word counts, this function computes the probability each word will appear if randomly selected from the input corpus.

````
def get_probs(word_count_dict):
    '''
    Input:
        word_count_dict: The wordcount dictionary where key is the word and value is its frequency.
    Output:
        probs: A dictionary where keys are the words and the values are the probability that a word will occur. 
    '''
    probs = {}
    probs = {k:float(v)/sum(word_count_dict.values()) for (k,v) in word_count_dict.items()}
    
    return probs
````

##N-grams

Contiguous sequence of n items from a given sample of text or speech

````
import nltk               # NLP toolkit
import re                 # Library for Regular expression operations

nltk.download('punkt')   
````

````
def sentence_to_trigram(tokenized_sentence):
    """
    Prints all trigrams in the given tokenized sentence.
    
    Args:
        tokenized_sentence: The words list.
    
    Returns:
        No output
    """
    # note that the last position of i is 3rd to the end
    for i in range(len(tokenized_sentence) - 3 + 1):
        # the sliding window starts at position i and contains 3 words
        trigram = tokenized_sentence[i : i + 3]
        print(trigram)

tokenized_sentence = ['i', 'am', 'happy', 'because', 'i', 'am', 'learning', '.']

print(f'List all trigrams of sentence: {tokenized_sentence}\n')
sentence_to_trigram(tokenized_sentence)
````

##Perplexity

This is a common metric to measure language models. It is related to entropy and it measures how random a text is. An arbitrary list of words will have a higher perplexity than a text generated by a human.


$PP(W) = P(s_{1}, s_{2}, s_{3}, ... , s_{m}) ^{(1/m)}$




Where,

W: Test set containing m sentences s

$s_{i}$: is the ith sentence in the set, each ending with < /s >

m: number of all words in in the set W including </s> but not < s >




The lower the perplexity score the better. It is more likely the sentence will sound natural. Good language models have a perplexity score between 20 and 60. 

The perplexity of a **bigram** model follows,

$PP(W) = P(\sqrt{\prod_{i=1}^m \prod_{j=1}^{\lvert s_{i} \rvert} \frac{1}{P(w_{j}^{(i)}\vert w_{j-1}^{(i)})} })$

##TensorFlow and Keras

TensorFlow is a deep learning library authored by Google. 

Let's work out a simple example. First we import the necessary libraries.

````
import tensorflow as tf
import numpy as np
from tensorflow import keras # Keras is built on top of TensorFlow
````

Lets work with some dummy data. 

````
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
````

````
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
````
